# Youtube Data Analysis 

This dataset contains the youtube data for country india and Data is provided in between 2017 and 2018.

Dataset fields include video_id, trending_date, title, channel_title, category_id,publish_time, tags, views, likes, dislikes, comment_count.
There are total 37352 rows and 16 columns

# Data Preparation and Cleaning

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

* Read the YouTube india csv_file using pandas 

```python
video_df = pd.read_csv("INvideos.csv")
video_df
```

Here's what we can tell by looking at the above dataframe:

- There are  37352 rows and 16 columns.
- Data is provided in between 2017 and 2018.

By using `info()` method. we can check the `Datatype` , `Column_name` and `count` of values.


```python
video_df.info()
```

We can see that there are `37,352` entries in the dataset. We can see also that all columns in the dataset are complete (i.e. they have `37,352` non-null entries) except description column which has some null values; it only has `36791` non-null values

## Data cleaning
The description column has some null values. These are some of the rows whose description values are null. We can see that null values are denoted by NaN


```python
video_df[video_df["description"].apply(lambda x: pd.isna(x))].head()
```

So to do some sort of data cleaning, and to get rid of those null values, we put an `empty` string in place of each null value in the `description` column

```python
video_df['description'] = video_df['description'].fillna(value = "")
```

```python
video_df.info()
```

Formating `trending_date` and `publish_date` columns. `pd.to_datetime` will convert the column datatype to datetime

```python
video_df['trending_date'] = pd.to_datetime(video_df['trending_date'], format='%y.%d.%m')
```

```python

video_df['publish_time'] = pd.to_datetime(video_df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
video_df.info()
```

#### Convert Rate Distribution

I will create some new variables to us analyse the` percent of likes, dislikes, comment` by each category to discover what category have the highest engagemen

```python
like_rate = video_df['likes'] /video_df['views'] *100

dislikes_rate = video_df['dislikes'] /video_df['views'] *100 

comment_rate = video_df['comment_count'] /video_df['views'] *100
```

```python
plt.figure(figsize = (13,6))

g1 = sns.distplot(dislikes_rate, color='red',hist=False, label="Dislike")
g1 = sns.distplot(like_rate, color='green',hist=False, label="Like")
g1 = sns.distplot(comment_rate,hist=False,label="Comment")
g1.set_title('CONVERT RATE DISTRIBUITION', fontsize=16)
plt.legend(fontsize=18)
plt.show()
```

#### What Publishing Time Receives the Most Views ?

```python
video_df['hour'] = video_df['publish_time'].dt.hour
by_hour = video_df.groupby('hour').mean()

plt.figure(figsize=(12,6))
plt.plot(by_hour.index.values, by_hour["views"])
plt.scatter(by_hour.index.values, by_hour["views"])

plt.xlabel("Hour of Day", fontsize=20)
plt.ylabel("Average Number of Views", fontsize=20)
plt.title("Average Amount of Views on Trending Videos by the Hour" , fontsize=20)
plt.show()
```
